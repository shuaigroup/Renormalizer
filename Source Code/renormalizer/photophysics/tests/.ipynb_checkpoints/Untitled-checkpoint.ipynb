```python
from renormalizer.utils.constant import *
import numpy as np
import scipy.fftpack as fft
import matplotlib.pyplot as plt
```

```python
t = np.load("autocorr_time.npy")
print(len(t), t[1]-t[0])
corrfunc = np.load("autocorr.npy")
corrfunc *= np.exp(-0.02*ev2au  * t)
plt.plot(t, corrfunc.real)
plt.plot(t, corrfunc.imag)
```

```python
y1 = fft.fftshift(fft.ifft(corrfunc.imag*2))*len(t)*(t[1]-t[0])   # +i\omega t backward
#y1 = fft.fftshift(fft.fft(func(t)))*(t[1]-t[0])   # -i\omega t forward
x1 = fft.fftshift(fft.fftfreq(len(t),t[2]-t[1]))*2*np.pi
print(x1[1]-x1[0])
#y1 += y1.conj()  # the -\infty to 0

#plt.plot(x1*au2ev, np.real(y1),"o-",label="real")
plt.plot(x1*au2ev, -np.imag(y1),"o-",label="imag")

plt.xlim(2.7,4)
plt.ylim(-1,1000)
plt.legend()
plt.show()
```
