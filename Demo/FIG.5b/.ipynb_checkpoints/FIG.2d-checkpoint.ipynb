```python
#2d
length = 36


_lambda = 0.012757738708152964 *0.25
w0 = 1400/au2cm
d0 = np.sqrt(_lambda*2 / (w0**2))
e_ad = 10*_lambda

e_ph_list=[]
e_ph_list.append('../data/2d_zt/2d_j_0lambda-1p1a/test.log')
e_ph_list.append('../data/2d_zt/2d_j_0.1lambda-1p1a/test.log')
e_ph_list.append('../data/2d_zt/2d_j_0.2lambda-1p1a/test.log')
e_ph_list.append('../data/2d_zt/2d_j_0.3lambda-1p1a/test.log')
e_ph_list.append('../data/2d_zt/2d_j_0.4lambda-1p1a/test.log')
e_ph_list.append('../data/2d_zt/2d_j_0.5lambda-1p1a/test.log')
e_ph_list.append('../data/2d_zt/2d_j_0.6lambda-1p1a/test.log')
e_ph_list.append('../data/2d_zt/2d_j_0.7lambda-1p1a/test.log')
agg_e_list_2d= np.zeros(len(e_ph_list))
for istate in range(len(e_ph_list)):
    with open(e_ph_list[istate],'r') as f:
        for lines in f:
            split_line = lines.rstrip().split()
            if len(split_line) >= 9: 
                if split_line[6] == 'lowest':
                    agg_e_list_2d[istate] = float(split_line[9])
                    break


    


e_ph_list=[]
e_ph_list.append(np.load('../data/2d_zt/2d_j_0lambda-1p1a/agg_s1_e_ph_correlation.npy'))
e_ph_list.append(np.load('../data/2d_zt/2d_j_0.1lambda-1p1a/agg_s1_e_ph_correlation.npy'))
e_ph_list.append(np.load('../data/2d_zt/2d_j_0.2lambda-1p1a/agg_s1_e_ph_correlation.npy'))
e_ph_list.append(np.load('../data/2d_zt/2d_j_0.3lambda-1p1a/agg_s1_e_ph_correlation.npy'))
e_ph_list.append(np.load('../data/2d_zt/2d_j_0.4lambda-1p1a/agg_s1_e_ph_correlation.npy'))
e_ph_list.append(np.load('../data/2d_zt/2d_j_0.5lambda-1p1a/agg_s1_e_ph_correlation.npy'))
e_ph_list.append(np.load('../data/2d_zt/2d_j_0.6lambda-1p1a/agg_s1_e_ph_correlation.npy'))
e_ph_list.append(np.load('../data/2d_zt/2d_j_0.7lambda-1p1a/agg_s1_e_ph_correlation.npy'))


lattice = 6
x=np.linspace(-20,20,20*2-1)
dx=0.5
dy=0.5
x = np.arange(-2.0, 2.5, dx)
y = np.arange(-2.0, 2.5, dy)

agg_lambda_2d=[]
extent = np.min(x), np.max(x), np.min(y), np.max(y)
for inum  in range(len(e_ph_list)):
    sum_d_mt = np.zeros((lattice,lattice))
    for i in range(lattice**2):
        for j in range(lattice**2):
            if np.floor(j/lattice) >=  np.floor(i/lattice):
                icolumn=int(np.floor(j/lattice) - np.floor(i/lattice))
            else:
                icolumn=int(np.floor(j/lattice) - np.floor(i/lattice))+lattice
            if j%lattice  >= i%lattice:
                iline = j%lattice - i%lattice
            else:
                iline = j%lattice - i%lattice +lattice            
            sum_d_mt[icolumn,iline] = sum_d_mt[icolumn,iline] + e_ph_list[inum][1,i,j]

        
    x= 2
    y=2
    center= tuple([x,y])
    tuple_x = (np.arange(lattice) -x)% lattice 
    tuple_y = (np.arange(lattice) -y)% lattice 
    res_sum_d_mt = (sum_d_mt[tuple_x,:])[:,tuple_y]
    plt.figure(figsize=(12,9))
    plt.imshow(np.abs(res_sum_d_mt[0:-1,0:-1])/d0,cmap=cm.pink,vmax=1,vmin=0,extent=extent)
    plt.yticks([-2,-1,0,1,2])
    plt.xticks([-2,-1,0,1,2])
    plt.xlabel('j=%.2f' % (inum*0.1))
    plt.colorbar()
    agg_lambda_2d.append(np.sum(np.abs(res_sum_d_mt[0:-1,0:-1])**2/d0**2))

print(e_ph_list[1].shape)
np.savez("2d_zt_ree_e",ree=agg_lambda_2d,e=agg_e_list_2d)


```
