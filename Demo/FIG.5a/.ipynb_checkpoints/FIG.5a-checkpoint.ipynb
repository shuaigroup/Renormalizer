```python
a_list=[]

a_list.append(np.load('../data/Dimer_zt/dimer_j_0lambda-1p1a/test.npz')['ACF'])
a_list.append(np.load('../data/Dimer_zt/dimer_j_0.1lambda-1p1a/test.npz')['ACF'])
a_list.append(np.load('../data/Dimer_zt/dimer_j_0.2lambda-1p1a/test.npz')['ACF'])
a_list.append(np.load('../data/Dimer_zt/dimer_j_0.3lambda-1p1a/test.npz')['ACF'])
a_list.append(np.load('../data/Dimer_zt/dimer_j_0.4lambda-1p1a/test.npz')['ACF'])
a_list.append(np.load('../data/Dimer_zt/dimer_j_0.5lambda-1p1a/test.npz')['ACF'])
a_list.append(np.load('../data/Dimer_zt/dimer_j_0.6lambda-1p1a/test.npz')['ACF'])
a_list.append(np.load('../data/Dimer_zt/dimer_j_0.7lambda-1p1a/test.npz')['ACF'])
a_list.append(np.load('../data/Dimer_zt/dimer_j_0.8lambda-1p1a/test.npz')['ACF'])
a_list.append(np.load('../data/Dimer_zt/dimer_j_0.9lambda-1p1a/test.npz')['ACF'])
a_list.append(np.load('../data/Dimer_zt/dimer_j_1.0lambda-1p1a/test.npz')['ACF'])
a_list.append(np.load('../data/Dimer_zt/dimer_j_1.1lambda-1p1a/test.npz')['ACF'])
a_list.append(np.load('../data/Dimer_zt/dimer_j_1.2lambda-1p1a/test.npz')['ACF'])
a_list.append(np.load('../data/Dimer_zt/dimer_j_1.3lambda-1p1a/test.npz')['ACF'])
a_list.append(np.load('../data/Dimer_zt/dimer_j_1.4lambda-1p1a/test.npz')['ACF'])
a_list.append(np.load('../data/Dimer_zt/dimer_j_1.5lambda-1p1a/test.npz')['ACF'])
a_length = len(a_list)



_lambda = 0.012757738708152964 *0.25
e_ad = _lambda *10
e_list =np.linspace(-1*e_ad,27*e_ad,8000)

spectra_list =[]

set_ead = _lambda *20
ic_list= []

for var in range(0,a_length):
    print('var is %d' % var)
    a_orignal = a_list[var].copy()
    a = np.array(a_orignal.copy())
    step=30
    dt = 0.2*100/2.4188  



    
    nsteps = int(len(a))
    print('nsteps used %d' % nsteps)

    n = 10 *8
    
    w= 0.0001*n


    num =np.array([dt*i for i in range(nsteps)])




          
    spectra_list.append(np.array([sci.trapz(a*np.exp(num*1j*(e-0))*np.exp(-(w**2)*(num**2)/2),x=num) for e in e_list]).real)
    e0= set_ead -e_ad
    ic_list.append(sci.trapz(a*np.exp(num*1j*(e0-0))*np.exp(-(w**2)*(num**2)/2),x=num).real)
    print(var,(sci.cumtrapz(a*np.exp(num*1j*(e0-0))*np.exp(-(w**2)*(num**2)/2),x=num).real)[-10:]*2*(1/2.4188)*1e17)

plt.figure(figsize=(12,9))
plt.plot(np.arange(len(a_list))*0.1,ic_list/ic_list[0])
np.save('Dimer_zt',ic_list/ic_list[0])
```
